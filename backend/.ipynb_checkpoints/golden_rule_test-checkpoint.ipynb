```python
import logging
import asyncio
import json
from app.core.qdrant import qdrant_client
from app.core.voyage import embed_text
# pyrefly: ignore [missing-import]
import nest_asyncio
def load_golden_set() -> list[dict]:
    with open("golden_set.json", "r", encoding="utf-8") as file:
        return json.load(file)

RAG_COLLECTION_NAME = "ironchat_docs"

async def _run_rag_retrieval(question: str, top_k: int = 3) -> list[str]:
    """
    Embed the user's question, search Qdrant for the closest document
    chunks, and return their raw text. Returns [] if nothing useful found.
    """
    try:
        question_vector = await asyncio.to_thread(embed_text, question)

        results = await qdrant_client.query_points(
            collection_name=RAG_COLLECTION_NAME,
            query=question_vector,
            limit=top_k,
        )

        return results.points
    except Exception as e:
        logging.error(f"RAG retrieval failed: {e}")
        return []

async def main():
    golden_set = load_golden_set()
    item = golden_set[0]
    query = item["query"]
    print(f"Testing Query: {query}")
    
    points = await _run_rag_retrieval(query)
    print(f"Retrieved {len(points)} chunks.")
    if points:
        for i, point in enumerate(points):
            print(f"\n--- Point {i+1} ---")
            print("Point ID:", getattr(point, "id", None))
            print("Score:", getattr(point, "score", None))
            print("Payload:", getattr(point, "payload", None))
    
    retrieved_ids = [point.id for point in points]
    relevant_ids = set(item["relevant_chunk_ids"])
    
    relevant_retrieved = sum(
        1 for chunk_id in retrieved_ids
        if chunk_id in relevant_ids
    )
     
    
    precision_at_k = relevant_retrieved / len(retrieved_ids)
    
    recall_at_k = (
            relevant_retrieved / len(relevant_ids)
            if relevant_ids
            else 0.0
    )
    reciprocal_rank = 0.0

    for rank, chunk_id in enumerate(retrieved_ids, start=1):
        if chunk_id in relevant_ids:
            reciprocal_rank = 1 / rank
            break

    
    print("Retrieved IDs:", retrieved_ids)
    print("Relevant IDs:", relevant_ids)
    print("Precision@3:", precision_at_k)
    print("Recall@3:", recall_at_k)
    print("Reciprocal Rank:", reciprocal_rank)
    
if __name__ == "__main__":
    try:
        loop = asyncio.get_running_loop()
    except RuntimeError:
        loop = None

    if loop and loop.is_running():

        nest_asyncio.apply()
        asyncio.run(main())
    else:
        asyncio.run(main())
```
