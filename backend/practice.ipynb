```python
ground_truth = {"doc_1","doc_2","doc_5",}
retrived = {"doc_1","doc_2","doc_3",}

true_positives = ground_truth & retrived

true_positives
```

```python
precision = len(true_positives) / len(retrived)
```

```python
precision
```

```python
recall = len(true_positives) / len(ground_truth)
```

```python
recall
```

```python
F1 = 2 * (precision * recall) / (precision + recall)
F1
```
